## Ćwiczenie 1: Series — tworzenie i operacje

```python
import pandas as pd
import numpy as np

# Series z listy — domyślny indeks (0, 1, 2, ...)
oceny = pd.Series([4.5, 3.0, 5.0, 3.5, 4.0])
print(oceny)
```

Zadanie 1a: Utwórz Series pensje z dict:

```python
pensje = pd.Series({
    'Anna': 5500,
    'Jan': 7200,
    'Ewa': 4800,
    'Marek': 9100,
    'Kasia': 6300
})

print(pensje)
```

Zadanie 1b: Odpowiedz na pytania o pensje

1. Wyświetl imię osoby zarabiającej najwięcej (użyj idxmax())


```python
najlepszy = pensje.idxmax()
print("Najwięcej zarabia:", najlepszy)
```

2. Oblicz średnią pensję

```python
srednia = pensje.mean()
print("Średnia pensja:", srednia)
```

3. Wyświetl imiona osób zarabiających powyżej średniej (filtrowanie boolean)

```python
powyzej = pensje[pensje > srednia]

print("Powyżej średniej:")
print(powyzej)
```

4. Policz ile osób zarabia poniżej 6000 (użyj len() lub .sum())

```python
ponizej_6000 = (pensje < 6000).sum()

print("Ile osób < 6000:", ponizej_6000)
```

Zadanie 1c: Oblicz pensje po zmianach:

1. Oblicz pensje po podwyżce 10% dla wszystkich (pensje * 1.10)

```python
podwyzka = pensje * 1.10
print("Po podwyżce:")
print(podwyzka)
```

2. Oblicz pensje po premii 500 zł dla wszystkich (pensje + 500)

```python
premia = pensje + 500
print("Po premii:")
print(premia)
```

3. Oblicz nową pensję = (pensja * 1.10) + 500

```python
nowe_pensje = (pensje * 1.10) + 500

print("Nowe pensje:")
print(nowe_pensje)
```

4. Wyświetl ranking nowych pensji od najwyższej (użyj sort_values)

```python
ranking = nowe_pensje.sort_values(ascending=False)

print("Ranking nowych pensji:")
print(ranking)
```

### Zadanie 3d: Brakujące dane

1. Policz brakujące wartości w każdej kolumnie (isna().sum())

```python
print(penguins.isna().sum())
```

2. Jaki procent danych brakuje w kolumnie 'sex'?

```python
braki = penguins['sex'].isna().sum()
procent = (braki / len(penguins)) * 100

print("Braki w sex (%):", procent)
```

## Ćwiczenie 4: Pytania biznesowe

```python
print(tips['day'].value_counts())
```

4. Policz ile rachunków jest od palaczy (value_counts na 'smoker')

```python
print(tips['smoker'].value_counts())
```

5. Wyświetl describe() tylko dla kolumny 'tip' — jaka jest mediana napiwku?

```python
print(tips['tip'].describe())
print("Mediana tip:", tips['tip'].median())
```
