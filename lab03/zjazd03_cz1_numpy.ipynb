## Ćwiczenie 3: Operacje wektorowe — analiza biznesowa 

```python
import numpy as np

# Produkty w sklepie internetowym
nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Ładowarka',
         'Klawiatura', 'Mysz', 'Monitor', 'Kamera', 'Drukarka']

ceny = np.array([3499, 1899, 2499, 349, 89, 249, 129, 1499, 599, 899])
ilosci_magazyn = np.array([15, 42, 78, 156, 312, 89, 201, 23, 45, 34])
oceny_klientow = np.array([4.7, 4.2, 4.5, 4.8, 3.9, 4.1, 4.3, 4.6, 3.8, 3.5])
```

1. Podstawowe statystyki cenowe 

```python
print(f'''Średnia cena: {ceny.mean()},
Najdroższy produkt: {nazwy[ceny.argmax()]} {ceny.max()} ,
Najtańszy produkt: {nazwy[ceny.argmin()]} {ceny.min()}
''')
```

2. Wartość magazynu

```python
wartosc_produktu = ceny * ilosci_magazyn

j = 0
for i in wartosc_produktu:
    print(nazwy[j], ":", wartosc_produktu[j])
    j += 1


print(f'Łączna wartość magazynu: {wartosc_produktu.sum()}')
```

3. Analiza cenowa

```python
print("Ceny po rabacie 15% (promocja świąteczna):")
rabat_15 = ceny * 0.85

j = 0
for i in rabat_15:
    print(nazwy[j], ":", rabat_15[j])
    j += 1

print("\nCeny z VAT 23%:")
VAT = ceny * 1.23

j = 0
for i in VAT:
    print(nazwy[j], ":", VAT[j])
    j += 1


produkty_1000 = (ceny > 1000).sum()
print('\nProdukty kosztujące powyżej 1000 zł:', produkty_1000)

```

4. Filtrowanie

```python
print("Produkty z oceną >= 4.5:")

maska_oceny = oceny_klientow >= 4.5

j = 0
for i in maska_oceny:
    if i:
        print(nazwy[j], ":", ceny[j])
    j += 1

print("\nProdukty poniżej 200 zł:")

maska_ceny = ceny < 200

j = 0
for i in maska_ceny:
    if i:
        print(nazwy[j], ":", ilosci_magazyn[j])
    j += 1
```

5. Analiza zapasów

```python
print("Produkty z niskim stanem magazynowym (<30):")

maska_magazyn = ilosci_magazyn < 30

j = 0
for i in maska_magazyn:
    if i:
        print(nazwy[j])
    j += 1

wartosci = ceny * ilosci_magazyn
suma = 0

j = 0
for i in maska_magazyn:
    if i:
        suma += wartosci[j]
    j += 1

print("\nŁączna wartość:", suma)

total = wartosci.sum()
procent = (suma / total) * 100

print("\nProcent magazynu:", procent, "%")
```

6. Analiza korelacji cena-ocena

```python
print("Produkty - okazje:")

maska = (oceny_klientow >= 4.5) & (ceny < 1000)

j = 0
for i in maska:
    if i:
        print(nazwy[j], ":", ceny[j])
    j += 1

maska_high = oceny_klientow >= 4.0

suma = 0
count = 0

j = 0
for i in maska_high:
    if i:
        suma += ceny[j]
        count += 1
    j += 1

avg_high = suma / count

maska_low = oceny_klientow < 4.0

suma = 0
count = 0

j = 0
for i in maska_low:
    if i:
        suma += ceny[j]
        count += 1
    j += 1

avg_low = suma / count

print("\nŚrednia cena (>= 4.0):", avg_high)
print("Średnia cena (< 4.0):", avg_low)
```

## Ćwiczenie 4: Benchmark — lista vs NumPy

```python
import time

# Przygotowanie danych
rozmiar = 1_000_000
lista = list(range(rozmiar))
tablica = np.array(lista)

# Test 1: Pomnóż przez 2
start = time.perf_counter()
wynik_lista = [x * 2 for x in lista]
czas_lista = time.perf_counter() - start

start = time.perf_counter()
wynik_numpy = tablica * 2
czas_numpy = time.perf_counter() - start

print(f"=== Mnożenie ×2 ({rozmiar:,} elementów) ===")
print(f"Lista:  {czas_lista*1000:.1f} ms")
print(f"NumPy:  {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
# Test 2: Suma
start = time.perf_counter()
suma_lista = sum(lista)
czas_lista = time.perf_counter() - start

start = time.perf_counter()
suma_numpy = tablica.sum()
czas_numpy = time.perf_counter() - start

print(f"\n=== Suma ({rozmiar:,} elementów) ===")
print(f"Lista:  {czas_lista*1000:.1f} ms")
print(f"NumPy:  {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
# Test 3: Filtrowanie (elementy > 500000)
start = time.perf_counter()
filtr_lista = [x for x in lista if x > 500000]
czas_lista = time.perf_counter() - start

start = time.perf_counter()
filtr_numpy = tablica[tablica > 500000]
czas_numpy = time.perf_counter() - start

print(f"\n=== Filtrowanie > 500k ({rozmiar:,} elementów) ===")
print(f"Lista:  {czas_lista*1000:.1f} ms")
print(f"NumPy:  {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```
